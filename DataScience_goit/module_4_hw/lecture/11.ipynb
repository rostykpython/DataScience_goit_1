```python
from sympy import var, plot_implicit, Eq, plot
x, y = var('x y')


a = 2

plot_implicit(Eq(x ** 3 + y ** 3, 3 * a * x * y))
```

```python
x, y = var('x y')
a = 1
b = 1
c = 1
plot_implicit(Eq((x ** 2 + y ** 2 - 2 * a * x) ** 2, b **2 * (x ** 2 + y ** 2) + c))
```
