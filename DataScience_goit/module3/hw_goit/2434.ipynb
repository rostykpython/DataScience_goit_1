```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
x0 = np.linspace(0, 100, 100)
y0 =  x0 * 10 + 2
```

```python
x1 = np.linspace(0, 100, 100)
y1 = x1 * 10 + 2
noise = np.random.normal(0, 30, size=y1.size)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x0, y0, x1, y1+noise, 'o')
```

```python
X = [(x, 1) for x in x1]
```

```python
def error(w, X, Y):
    err = Y - np.matmul(X, w)
    return err.dot(err)
```

```python
result = optimize.minimize(
    error,
    (0, 1),
    args=(X, y1 + noise)

)

w0, w1 = result.x
y2 = w0 * x0 + w1
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x0, y2, 'o', x0, y0)
```
