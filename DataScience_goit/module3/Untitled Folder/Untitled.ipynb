```python
t = np.linspace(0, 4*np.pi, 100)

y1 = np.sin(t)
y2 = 2 * np.sin(2 * t + np.pi / 4)
```

```python
plt.plot(t, y1, '-', label='s')
plt.plot(t, y2, '--', label='ss')


plt.xlabel('t')
plt.ylabel('y')
plt.title('')
plt.legend()

plt.axhline(color='r')
plt.axvline(color='r')
```
